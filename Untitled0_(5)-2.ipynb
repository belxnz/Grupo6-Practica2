# **CONSTRUCCIN DE MODELOS**

Lo primero que hacemos antes de dividir los datos en un conjunto de train para entrenar y ajustar hiperparmetros y uno de test sobre el cual haremos las predicciones que nos seran utiles para obtener el porcentaje de precisin, es convertir los datos de las variables categoricas a numericas, para asi poder trabajar con ellas de manera correcta y uniforme al realizar el ajuste de hiperparametros. Esto lo haremos con la funcion get_dummies().

```python
from sklearn.model_selection import cross_val_score, TimeSeriesSplit
X = df.loc[:,df.columns != 'Attrition']
y = df['Attrition']

# Se convierten las variables categoricas en numericas utilizando one-hot encoding
X = pd.get_dummies(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=6)

tscv = TimeSeriesSplit(n_splits=5)
```

## **LogisticRegression**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.impute import SimpleImputer
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import balanced_accuracy_score, f1_score, confusion_matrix

# Se define el pipeline

start = time.time()

LogReg = LogisticRegression(random_state=6, class_weight='balanced')
pipe_lr = Pipeline([('imputer', SimpleImputer(strategy='mean')), ('scl', StandardScaler()), ('clf', LogReg)])

# Creamos el stratified k-fold cross-validator
cv = StratifiedKFold(n_splits=5)

# Inicializamos listas en las que guardaremos las scores de cada fold
bal_acc_scores = []
f1_scores = []

for train_index, test_index in cv.split(X, y):
    # Dividimos los datos entre train y test
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Fit del pipeline a los datos de train
    pipe_lr.fit(X_train, y_train)

    # Hacemos predicciones en los datos de test
    y_pred = pipe_lr.predict(X_test)

    # Calculamos el balanced accuracy score y F1 score
    bal_acc = balanced_accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, pos_label='Yes')

    # Aadimos dichos scores a las listas que hemos creado
    bal_acc_scores.append(bal_acc)
    f1_scores.append(f1)

end = time.time()

# Printeamos los resultados
print('Average balanced accuracy:', sum(bal_acc_scores)/len(bal_acc_scores))
print('Average F1 score:', sum(f1_scores)/len(f1_scores))

# Calculamos la matriz de confusion utilizando el mejor modelo
best_index = bal_acc_scores.index(max(bal_acc_scores))
best_train_index, best_test_index = list(cv.split(X, y))[best_index]
X_train, X_test = X.iloc[best_train_index], X.iloc[best_test_index]
y_train, y_test = y.iloc[best_train_index], y.iloc[best_test_index]
pipe_lr.fit(X_train, y_train)
y_pred = pipe_lr.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrix:\n', cm)

print("Tiempo de ejecucin: {:.2f} segundos".format(end-start))
```

## **HistGradientBoosting**

```python
from sklearn.ensemble import HistGradientBoostingClassifier

start = time.time()

# Define el pipeline
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('clf', HistGradientBoostingClassifier())
])

# Define los hiperparametros
param_grid = {
    'clf__learning_rate': [0.1, 0.01],
    'clf__max_depth': [3, 5, 7],
    'clf__max_iter': [10, 50, 100, 150],
    'clf__min_samples_leaf': [1, 2, 3]
}

# Hacemos el GridSearchCV
cv = StratifiedKFold(n_splits=5)
grid_search = GridSearchCV(pipeline, param_grid=param_grid, cv=cv, scoring='accuracy', error_score='raise')

# Hacemos el Fit de los resultados del GridSearchCV sobre los datos train
grid_search.fit(X_train, y_train)

# Imprimimos el mejor hiperparametro y su correspondiente score
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

# Hacemos una prediccion sobre los datos de test
y_pred = grid_search.predict(X_test)

# Calculamos el balanced accuracy score y F1 score
bal_acc = balanced_accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, pos_label='Yes')

end = time.time()

# Calculamos la matriz de confusion
cm = confusion_matrix(y_test, y_pred)

# Print the evaluation metrics
print("Balanced accuracy:", bal_acc)
print("F1 score:", f1)
print("Confusion matrix:\n", cm)

print("Tiempo de ejecucin: {:.2f} segundos".format(end-start))
```

## **Conclusiones**

Vamos a empezar comparando el balance accuracy, F1 score y la matriz de confusin. Antes de realizar esta comparacin, creemos que es importante explicar qu significa cada uno de estos mtodos de evaluacin. El 'balance accuracy' se utiliza evaluar el rendimiento de un modelo de clasificacin, especialmente cuando las variables con las que trata estan desbalanceadas, ya que este mtodo le da el mismo peso a ambas variables independientemente de la diferencia en porcentajes de missing values que cada una posee. Por otra parte, 'F1 score' es un mtodo ms ideal para problemas de clasificacin balanceada, debido a que no realiza ningun tipo de ajuste de pesos.
Finalmente, la matriz de confusion es una tabla que contiene la cantidad de predicciones correctas (diagonal izq-der) e incorrectas (diagonal der-izq) realizadas por un modelo de clasificacin en funcin de las clases reales y las clases predichas.

## **LogisticRegression**

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Define el pipeline
LogReg = LogisticRegression(random_state=6, class_weight='balanced')
pipe_lr = Pipeline([('imputer', SimpleImputer(strategy='mean')), ('scl', StandardScaler()), ('select', SelectKBest(f_classif, k=10)), ('clf', LogReg)])

# Creamos el stratified k-fold cross-validator
cv = StratifiedKFold(n_splits=5)

# Inicializamos listas en las que guardaremos las scores de cada fold
bal_acc_scores = []
f1_scores = []


for train_index, test_index in cv.split(X, y):

    # Dividimos los datos entre train y tests
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Fit del pipeline a los datos de train
    pipe_lr.fit(X_train, y_train)

    # Hacemos predicciones en los datos de test
    y_pred = pipe_lr.predict(X_test)

    # Calculamos el balanced accuracy score y F1 score
    bal_acc = balanced_accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, pos_label='Yes')

    # Aadimos dichos scores a las listas que hemos creado
    bal_acc_scores.append(bal_acc)
    f1_scores.append(f1)

# Printeamos los resultados
print('Average balanced accuracy:', sum(bal_acc_scores)/len(bal_acc_scores))
print('Average F1 score:', sum(f1_scores)/len(f1_scores))

# Calculamos la matriz de confusion utilizando el mejor modelo
best_index = bal_acc_scores.index(max(bal_acc_scores))
best_train_index, best_test_index = list(cv.split(X, y))[best_index]
X_train, X_test = X.iloc[best_train_index], X.iloc[best_test_index]
y_train, y_test = y.iloc[best_train_index], y.iloc[best_test_index]
pipe_lr.fit(X_train, y_train)
y_pred = pipe_lr.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrix:\n', cm)
```

```python
#Creamos el pipeline aadiendo un mtodo de seleccin de atributos de tipo filter(f_classif)
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('selector', SelectKBest(f_classif, k=5)),
    ('clf', HistGradientBoostingClassifier())
])

param_grid = {
    'clf__learning_rate': [0.1, 0.01],
    'clf__max_depth': [3, 5, 7],
    'clf__max_iter': [10, 50, 100, 150],
    'clf__min_samples_leaf': [1, 2, 3]
}

cv = StratifiedKFold(n_splits=5)
grid_search = GridSearchCV(pipeline, param_grid=param_grid, cv=cv, scoring='balanced_accuracy', error_score='raise')

# Fit sobre los resultados del GridSearchCV
grid_search.fit(X_train, y_train)

# Print de los mejores hiperparmetros
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

# Predecimos sobre los datos de testing
y_pred = grid_search.predict(X_test)

# Calculamos el balanced accuracy score y F1 score
bal_acc = balanced_accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, pos_label='Yes')

# Calculamos la matriz de confusion
cm = confusion_matrix(y_test, y_pred)

# Printeamos los resultados
print("Balanced accuracy:", bal_acc)
print("F1 score:", f1)
print("Confusion matrix:\n", cm)

# Printeamos los atributos ms importantes
selector = grid_search.best_estimator_.named_steps['selector']
selected_features = selector.get_support(indices=True)
all_features = X.columns
important_features = all_features[selected_features]
print("Most important features:", important_features)
```

```python
from lightgbm import LGBMClassifier

start = time.time()

# Define el pipeline
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('clf', LGBMClassifier())
])

# Define los hiperparametros
param_grid = {
    'clf__n_estimators': [50, 100, 200],
    'clf__max_depth': [3, 5, 7],
    'clf__learning_rate': [0.05, 0.1, 0.2]
}

# Hacemos el GridSearchCV
cv = StratifiedKFold(n_splits=5)
grid_search = GridSearchCV(pipeline, param_grid=param_grid, cv=cv, scoring='accuracy', error_score='raise')

# Hacemos el Fit de los resultados del GridSearchCV sobre los datos train
grid_search.fit(X_train, y_train)

# Imprimimos el mejor hiperparametro y su correspondiente score
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

# Hacemos una prediccion sobre los datos de test
y_pred = grid_search.predict(X_test)

# Calculamos el balanced accuracy score y F1 score
bal_acc = balanced_accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, pos_label='Yes')

end = time.time()

# Calculamos la matriz de confusion
cm = confusion_matrix(y_test, y_pred)

# Print the evaluation metrics
print("Balanced accuracy:", bal_acc)
print("F1 score:", f1)
print("Confusion matrix:\n", cm)

print("Tiempo de ejecucin: {:.2f} segundos".format(end-start))
```

```python
best_model = grid_search.best_estimator_
with open('/content/final_model.pkl', 'wb') as f:
    pickle.dump(best_model, f)
```
