## Tipo de problema


```python
X = df.loc[:,df.columns != 'Attrition']
y = df['Attrition']

# Preprocess convert categorical variables to numerical variables using one-hot encoding
X = pd.get_dummies(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=6)
```

LogicRegression

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.impute import SimpleImputer
from sklearn.metrics import balanced_accuracy_score, f1_score, confusion_matrix

# Define the pipeline
LogReg = LogisticRegression(random_state=6, class_weight='balanced')
pipe_lr = Pipeline([('imputer', SimpleImputer(strategy='mean')), ('scl', StandardScaler()), ('clf', LogReg)])

# Set up the stratified k-fold cross-validator
cv = StratifiedKFold(n_splits=5)

# Initialize lists to store the scores for each fold
bal_acc_scores = []
f1_scores = []

# Iterate over the folds
for train_index, test_index in cv.split(X, y):
    # Split the data into training and testing sets
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Fit the pipeline to the training data
    pipe_lr.fit(X_train, y_train)

    # Predict on the testing data
    y_pred = pipe_lr.predict(X_test)

    # Calculate the balanced accuracy score and F1 score
    bal_acc = balanced_accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, pos_label='Yes')

    # Append the scores to the lists
    bal_acc_scores.append(bal_acc)
    f1_scores.append(f1)

# Print the average scores across all folds
print('Average balanced accuracy:', sum(bal_acc_scores)/len(bal_acc_scores))
print('Average F1 score:', sum(f1_scores)/len(f1_scores))

# Calculate the confusion matrix using the best model
best_index = bal_acc_scores.index(max(bal_acc_scores))
best_train_index, best_test_index = list(cv.split(X, y))[best_index]
X_train, X_test = X.iloc[best_train_index], X.iloc[best_test_index]
y_train, y_test = y.iloc[best_train_index], y.iloc[best_test_index]
pipe_lr.fit(X_train, y_train)
y_pred = pipe_lr.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrix:\n', cm)
```

Boosting

```python
from sklearn.metrics import balanced_accuracy_score, f1_score, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.impute import SimpleImputer
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# Define the pipeline
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('clf', GradientBoostingClassifier())
])

# Define the hyperparameter space
param_grid = {
    'clf__learning_rate': [0.1, 0.01],
    'clf__n_estimators': [100, 200],
    'clf__max_depth': [3, 5, 7],
    'clf__min_samples_split': [2, 4],
    'clf__min_samples_leaf': [1, 2],
    'clf__max_features': ['auto', 'sqrt', 'log2']
}

# Set up the GridSearchCV object
cv = StratifiedKFold(n_splits=5)
grid_search = GridSearchCV(pipeline, param_grid=param_grid, cv=cv, scoring='accuracy', error_score='raise')

# Fit the GridSearchCV object to the data
grid_search.fit(X_train, y_train)

# Print the best hyperparameters and their corresponding score
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

# Predict on the testing data
y_pred = grid_search.predict(X_test)

# Calculate the balanced accuracy score and F1 score
bal_acc = balanced_accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, pos_label='Yes')

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Print the evaluation metrics
print("Balanced accuracy:", bal_acc)
print("F1 score:", f1)
print("Confusion matrix:\n", cm)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import balanced_accuracy_score, f1_score, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.impute import SimpleImputer
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# Define the pipeline
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('selector', SelectKBest(f_classif, k=5)),
    ('clf', GradientBoostingClassifier())
])

# Define the hyperparameter space
param_grid = {
    'clf__learning_rate': [0.1, 0.01],
    'clf__n_estimators': [100, 200],
    'clf__max_depth': [3, 5, 7],
    'clf__min_samples_split': [2, 4],
    'clf__min_samples_leaf': [1, 2],
    'clf__max_features': ['auto', 'sqrt', 'log2']
}

# Set up the GridSearchCV object
cv = StratifiedKFold(n_splits=5)
grid_search = GridSearchCV(pipeline, param_grid=param_grid, cv=cv, scoring='accuracy', error_score='raise')

# Fit the GridSearchCV object to the data
grid_search.fit(X_train, y_train)

# Print the best hyperparameters and their corresponding score
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

# Predict on the testing data
y_pred = grid_search.predict(X_test)

# Calculate the balanced accuracy score and F1 score
bal_acc = balanced_accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, pos_label='Yes')

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Print the evaluation metrics
print("Balanced accuracy:", bal_acc)
print("F1 score:", f1)
print("Confusion matrix:\n", cm)

# Print the most important features
selector = grid_search.best_estimator_.named_steps['selector']
selected_features = selector.get_support(indices=True)
all_features = X.columns
important_features = all_features[selected_features]
print("Most important features:", important_features)
```
